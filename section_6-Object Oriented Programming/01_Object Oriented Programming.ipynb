## class

User defined objects are created using the <code>class</code> keyword.

The class is a blueprint that defines the nature of a future object.

From classes we can construct instances. 

An instance is a specific object created from a particular class. 

By convention we give classes a name that starts with a capital letter.

For example, we can create a class called Dog. An attribute of a dog may be its breed or its name, while a method of a dog may be defined by a .bark() method which returns a sound.

## Attributes


```python

class Dog():

    #class object attribute
    #same for any instance of a class
    species = 'Manmmal' 
    
    def __init__(self,breed,name,spots):
        self.breed = breed
        self.name = name
        self.spots = spots 
```

```python
my_dog = Dog(breed = "lab",name = "sammy",spots = True)
```

```python
my_dog = Dog("lab","sammy",True)
```

```python
my_dog.species
```

## Methods

Methods are functions defined inside the body of a class.

They are used to perform operations with the attributes of our objects.

You can basically think of methods as functions acting on an Object that take the Object itself into account through its *self* argument.

```python

class Dog():
    species = 'Manmmal'
    
    def __init__(self,breed,name):
        self.breed = breed
        self.name = name

    #operation/action------> Methods 
    def bark(self):
        print("WOOF")
```

```python
my_dog = Dog("lab","sammy")
```

```python
my_dog.bark()

# notice when we .tab, next to bark we can see that it says funciton, 
# always add "()" when calling a function
```

```python
#methods use info about the object it self

class Dog():
    species = 'Manmmal'
    
    def __init__(self,breed,name):
        self.breed = breed
        self.name = name

    #operation/action
    def bark(self):
        print(f"WOOF!!! my name is {self.name}")
```

```python
my_dog = Dog("lab","sammy")
```

```python
my_dog.bark()
```

```python
my_dog2 = Dog()  #it expect parameters
```

```python
# methods can take some outside arguments

class Dog():
    species = 'Manmmal'
    
    def __init__(self,breed,name):
        self.breed = breed
        self.name = name

    #operation/action
    def bark(self,number):  
        print(f"WOOF!!! my name is {self.name} and number is {number}.")
        
    
```

### inheritance

```python
#inheritance
#from new classes using classes that are allready being defined
```

```python
# base class
class Animal():
    def __init__(self):
        print("animal created")

    #method
    def who_am_i(self):
        print("I am an animal")

    # another method
    def eat(self):
        print("Iam eating")
        
```

```python
myanimal = Animal()
```

```python
myanimal.eat()
```

```python
myanimal.who_am_i()
```

```python
# Dog is inheriting objecg form the animal class
# this class is knoen as derived class
class Dog(Animal):

    def __init__(self):
        Animal.__init__(self)  #we have created an instance of animal class
        print("dog created")

    def who_am_i(self):
        print("I am a Dog")  #we can always over ride the base class method,

    #we can also add on new methods
    def bark(self):
        print("WOFF!!")
    

```

```python
#our Dog call also inherite methods from the Animal class
mydog = Dog()
```

```python
# all the methods in Aniaml class are also available for Dog() call becuse of the Inheritance
mydog.eat()
```

```python
# we can always ocer ride the base class methods
mydog.who_am_i()
```

```python
mydog.bark()
```

### polymorphism

```python
# how diff object class can share the same name
# and those methods can be called form the same name 
#even tho there are variety of diff object available
```

```python
class Dog():
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name+" say Woff"
```

```python
class Cat():
    def __init__(self,name):
        self.name = name

    def speak(self):
        return self.name+" say meow"
```

```python
# here two class with same methods name, 
# each method is unique to that instance of the class
```

```python
niko = Dog("niko")
felix = Cat("felix")
```

```python
print(niko.speak())
print(felix.speak())
```

```python
#adding method to work with len

class Book():
    def __init__(self,title,author,pages):

        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):   #help to print user defined objects  
        return f"{self.title} by {self.author}"

    def __len__(self):
        return self.pages
```

```python
b = Book("a","b",100)
```

```python
# ino order for extra things to occur when you call delete,

#adding method to work with len

class Book():
    def __init__(self,title,author,pages):

        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):    #help to print user defined objects
        return f"{self.title} by {self.author}"

    def __len__(self):
        return self.pages

    def __del__(self):
        print("A book is been deleted")
```

```python
b = Book("a","b",100)
```
