```python
class Animal():
    def __init__(self):
        print("Animal created")

    def who_am_i(self):
        print("I am an animal")

    def eat(self):
        print("I am Eating")
```

```python
class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog created")

    def eat(self):
        print("im a dog")

    def bark(self):
        print("WOOF")
```

```python
myanimal = Dog()
```

```python
myanimal.bark()
```

```python
myanimal.who_am_i()
```

```python
class Book():
    def __init__(self,title,author,pages):
        self.title = title
        self.author = author
        self.pages = pages

    def __str__(self):
        return f"{self.title}, {self.author}"

    def __len__(self):
        return self.pages

    def __del__(self):
        print("book deleted")
```

```python
b = Book("python rock","marx",200)
```

```python
print(b)
```

```python
len(b)
```

```python
class Line():
    def __init__(self,cord1,cord2):
        self.cord1 = cord1
        self.cord2 = cord2

    def distance(self):
        x1,y1 = self.cord1  #tuple unpacking
        x2,y2 = self.cord2

        return ((x2-x1)**2 + (y2-y1)**2)**0.5

    def slop(self):
        x1,y1 = self.cord1  #tuple unpacking
        x2,y2 = self.cord2

        return (y2-y1)/(x2-x1)
        
        

        
        
```

```python
cord1 = (3,2)
cord2= (8,10)
```

```python
li = Line(cord1,cord2)
```

```python
li.distance()
```
