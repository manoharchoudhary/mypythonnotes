## Dictionaries: key value pairing

###### --object retrieved by key name
###### --unordered and can not be sorted

### syntex: {'key1':'value1','key2':'value2'}

```python
# constructing dictionaries

my_dict = {'key1':'value1','key2':'value2','key3':'value3'}
```

```python
my_dict
```

```python
# to retrive a value

my_dict['key2']
```

```python
# use cases

price_lookup = {'apple':2.99,'oranges':1.99,'milk':0.8}
```

```python
price_lookup['oranges']
```

```python
#dictionary can hold other data types as well 

dic = {'k1':123,'k2':[4,5,6],'k3':{'a':'edd','b':'banana','c':['app','web','cash']}}
```

```python
dic['k1']
```

```python
dic['k2']
```

```python
dic['k3']
```

```python
dic['k2'][1]
```

```python
dic['k3']['c']
```

```python
# grab web
dic['k3']['c'][1]
```

```python
lists = dic['k2']
lists
```

```python
d= {'k1':100,'k2':200,'k3':300}
```

```python
# adding new key:value pair

d['k4']=400
```

```python
d
```

```python
# over writing existing key- value

d['k1'] = 150
```

```python
d
```

```python
# display keys

d.keys()
```

```python
 # display values
d.values()
```
