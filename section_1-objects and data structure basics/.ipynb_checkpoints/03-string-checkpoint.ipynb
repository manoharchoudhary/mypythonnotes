#### slicing 

```python
mystring = 'abcdefghijkl'
```

###### mystring[start:stop:step]

```python
#start from index position 2 and go till the end

mystring[2:]
```

```python
# go all the way till index position 4 but do not include 4

mystring[:4]
```

```python
# grab def

mystring[3:6]
```

```python
mystring[::]
```

```python
#print character from string with step size of 2

mystring[::2]
```

```python
mystring[2::2]
```

```python
mystring[2:8:3]
```

reversing string 

```python
mystring[::-1]
```
