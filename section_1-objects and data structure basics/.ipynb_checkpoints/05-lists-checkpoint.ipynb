```python
alp_list = ['a','e','x','b','m']
num_list = [4,5,9,7,6,3,1]
```

```python
# sort method
alp_list.sort()
#inplace method does change the OG list
```

```python
alp_list
```

```python
# this is a inplace method which do not return any output, it actually changes the actual list

my_sorted_list = alp_list.sort()
```

#### Empty list

######  An empty list has  a boolean value of "False"

This means that if you use an empty list in an if statement or any other context that requires a boolean value, it will be treated as False. 

```python
list1 = []
```

```python
bool(list1)
```
