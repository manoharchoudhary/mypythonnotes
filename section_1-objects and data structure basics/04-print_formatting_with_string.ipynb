# string formatting for printing

### printing variable name inside a string is called string interpolation.

###  .format() method
syntax:
    
  #### 'sting here {} then also{}'.format('something1','something2')

```python
print("my name is {}".format('manohar'))
```

inserting string based on indexing  

```python

print('The {} {} {}'.format('fox','brown','quick'))
```

```python
print('The {2} {1} {0}'.format('fox','brown','quick'))
```

```python
print('The {0} {0} {0}'.format('fox','brown','quick'))
```

```python
print('The {q} {b} {f}'.format(f='fox',b='brown',q='quick'))
```

### Float formatting follows "{value:width.precisionf}"

value: result 

width : adds white space before the result

precision: values after decimal

```python
result = 2225265/9
```

```python
result
```

```python
print("the result was {}".format(result))
```

```python
print("the result was {r}".format(r = result))
```

```python
#precision of 3 and width is 1

print('the result was {r:1.3f}'.format(r=result))
```

```python
r = result
```

```python
#precision of 5 and width is 20

print('the result was {r:20.5f}'.format(r=r))
```

### f-string method

```python
name = 'marx'
```

```python
# .format method

print("my name is {}".format(name))
```

```python
# f-string

print(f"my name is {name}")
```
