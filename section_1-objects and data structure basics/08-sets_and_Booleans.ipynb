##### when two set are compared, it ignores the order, rather focus on content

```python
str1 = "The quick brown fox jumps over the lazy dog"
str2 = "a bCdefg hijklmnop marx qrstuvwxyz"
```

```python
str1 = str1.replace(" ","").lower()
str2 = str2.replace(" ","").lower()
```

```python
print(str1)
print(str2)
```

```python
set(str1) == set(str2)
```
