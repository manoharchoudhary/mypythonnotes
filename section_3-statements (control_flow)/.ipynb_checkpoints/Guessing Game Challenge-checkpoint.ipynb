# Guessing Game Challenge

Let's use `while` loops to create a guessing game.

The Challenge:

Write a program that picks a random integer from 1 to 100, and has players guess the number. The rules are:

1. If a player's guess is less than 1 or greater than 100, say "OUT OF BOUNDS"
2. On a player's first turn, if their guess is
 * within 10 of the number, return "WARM!"
 * further than 10 away from the number, return "COLD!"
3. On all subsequent turns, if a guess is 
 * closer to the number than the previous guess return "WARMER!"
 * farther from the number than the previous guess, return "COLDER!"
4. When the player's guess equals the number, tell them they've guessed correctly *and* how many guesses it took!

You can try this from scratch, or follow the steps outlined below. A separate Solution notebook has been provided. Good luck!

#### Write a `while` loop that compares the player's guess to our number. If the player guesses correctly, break from the loop. Otherwise, tell the player if they're warmer or colder, and continue asking for guesses.

Some hints:
* it may help to sketch out all possible combinations on paper first!
* you can use the `abs()` function to find the positive difference between two numbers
* if you append all new guesses to the list, then the previous guess is given as `guesses[-2]`

```python
from random import randint
random_num = randint(1,100)
guesses = []

print("WELCOME TO GUESS ME!")
print("I'm thinking of a number between 1 and 100")
print("If your guess is more than 10 away from my number, I'll tell you you're COLD")
print("If your guess is within 10 of my number, I'll tell you you're WARM")
print("If your guess is farther than your most recent guess, I'll say you're getting COLDER")
print("If your guess is closer than your most recent guess, I'll say you're getting WARMER")
print("LET'S PLAY!")
print('\n')

while True:
    
    user_guess = int(input('Enter number between 1 and 100: '))
    if user_guess>100 or user_guess<1:
        print("OUT OF BOUNDS")
        continue
        
    guesses.append(user_guess)

    if len(guesses) == 1:
        if abs(user_guess-random_num)<=10:
            print("WARM")
        else:
            print("COLD")
    elif user_guess==random_num:
        print("Congratulations! You've guessed the number correctly!")
        print(f'it took you {len(guesses)} guesses.')
        break
        
    else:
        if abs(user_guess-random_num)>abs(guesses[-2]-random_num):
            print("COLDER")
        else:
            print("WARMER")
        

    
        
```

```python
#GPT

import random

def guessing_game():
    # Generate a random integer between 1 and 100
    target = random.randint(1, 100)
    print("Welcome to the Guessing Game!")
    print("Try to guess the number I'm thinking of between 1 and 100.")
    print("Type 'exit' to quit the game.")

    # Initialize variables
    guesses = []
    
    while True:
        # Ask the player for their guess
        guess = input("Enter your guess: ")
        
        # Allow player to exit the game
        if guess.lower() == 'exit':
            print("Thanks for playing! Goodbye!")
            break
        
        # Validate input
        if not guess.isdigit():
            print("Please enter a valid number.")
            continue
        
        guess = int(guess)
        
        # Check if the guess is out of bounds
        if guess < 1 or guess > 100:
            print("OUT OF BOUNDS")
            continue
        
        # Add the guess to the list of guesses
        guesses.append(guess)
        
        # Check if the guess is correct
        if guess == target:
            print(f"Congratulations! You guessed the number {target} correctly in {len(guesses)} guesses.")
            break
        
        # Give feedback on the first guess
        if len(guesses) == 1:
            if abs(guess - target) <= 10:
                print("WARM!")
            else:
                print("COLD!")
        else:
            # Compare the current guess to the previous guess
            if abs(guess - target) < abs(guesses[-2] - target):
                print("WARMER!")
            else:
                print("COLDER!")

# Run the game
guessing_game()
```
