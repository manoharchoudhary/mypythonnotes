```python
mystring2 = 'world'
```

```python
# by list comprehension
# its just the compact way of writing code
# computional time is about the same

mylist = [item for item in mystring2]
# this is a flatenout for_loop

mylist
```

```python
mylist = [x for x in 'word']
mylist
```

```python
mylist = list(range(11))
mylist
```

```python
# by list comprehension

mylist = [num for num in range(11)]
mylist
```

```python
# square by list comprehension

mylist = [num**2 for num in range(11)]
mylist
```

###### If condition

```python


mylist = [num for num in range(0,11) if num%2==0]

mylist
```

```python
# adding if statement 

mylist = [num**2 for num in range(11) if num % 2==0]
mylist
```

```python
celcius = [0,10,20,34.5]
```

```python
# arithmetic operations
# convert celcius to fehrenheit

far = [((9/5)*temp + 32) for temp in celcius]
far
```

```python
# in normal for loop

far = []

for temp in celcius:
    far.append(((9/5)*temp)+32)
    
far
```

##### if else statement in list comprehension

```python


results = [x if x%2==0 else "odd" for x in range(11)]
results
```

##### nested loop 

```python
mylist = []
for x in [2,4,6]:
    for y in [1,10,100]:
        mylist.append(x*y)
mylist        
```

```python
# nested loop in list comprehension

mylist = [x*y for x in [2,4,6] for y in [1,10,100]]
mylist
```
