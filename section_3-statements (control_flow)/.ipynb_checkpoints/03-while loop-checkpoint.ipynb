```python
# while

x = 0

while x < 5:
    print(f'value of x is :{x}')
    x += 1
```

```python
# while

x = 0

while x < 5:
    print(f'value of x is :{x}')
    x += 1
    
else:
    print('x is not less than 5')
```

### break, continue, pass

#### break: break out of the current closest enclosing loop.
#### continue: goes to the top of the closest enclosing loop.
#### pass: does nothing at all.

```python
x = [1,2,3]
```

```python
#pass : do nothing

for item in x:
    pass   #place filler
```

```python
#continue 

mystring = 'sammy'

for letter in mystring:
     
    if letter == 'a':
        continue # go back to the loop
    print(letter)    
```

```python
#break 

mystring = 'sammy'

for letter in mystring:
    if letter == 'a':
        break      # break out of the loop
    print(letter)    
```

```python
# break 

x = 0

while x <  5:
    if x == 2:
        break
    print(x)
    x += 1
```
