 ### 1. range function

```python
mylist = [1,2,3,4,5,6]
```

```python
mylist1 = list(range(1,10))
mylist1
```

```python
# range(start, stop, step size) 

for num in range(10): #go upto but not include 10
    print(num)
```

```python
for num in range(3,10):
    print(num)
```

```python
for num in range(0,11,2):
    print(num)
```

```python
range(0,11,2) #range is a generator it do not return anything

list(range(0,11,2)) #to have range function to return something; cast it into a list
```

### 2. Enumerate

```python
# eg program

index_count = 0

for letters in "abcde":
    print(f"at index {index_count}, letter is {letters}")
    index_count += 1
    
    #this is such a common fucntion that we will use enumerate to do the same 
```

```python
word = 'abcde'

for item in enumerate(word):
    print(item)
    
    #it return tuple with index count being the other attribute
    
    # with this we can use tuple unpacking
```

```python

word = 'abcde'

for index,letter in enumerate(word):
    print(index,letter)
    
#it return a counter    
```

```python
word = 'abcde'

for index,letter in enumerate(word):
    print(index)
    
#it return a counter    
```

```python
word = 'abcde'

for index,letter in enumerate(word):
    print(letter)
    
#it return a counter    
```

### 3. zip function  

#### kind of opposite to enumerate, it zips together two lists

```python
mylist1 = [1,2,3]
mylist2 = ['a','b','c']
```

```python
for item in zip(mylist1,mylist2):
    print(item)
    
    
```

```python
mylist3 = [100,200,300,400,500]
```

```python
for item in zip(mylist1,mylist2,mylist3):
    print(item)
    
    #if the lists are uneven in len then the zip funtion will go too the shortest list
```

```python
# again we can use tuple unpacking
for a,b,c in zip(mylist1,mylist2,mylist3):
    print(a,b,c)
```

```python

for a,b,c in zip(mylist1,mylist2,mylist3):
    print(a)
```

```python
zip(mylist1,mylist2,mylist3) #zip is a generator so it do not return anything 
```

```python
list(zip(mylist1,mylist2,mylist3)) #for it to return we have to cast it to a list
```
