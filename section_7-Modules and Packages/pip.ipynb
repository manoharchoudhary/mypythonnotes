#### pypi with pip install

pypi is a repository for open-source third-party python packages.

so far we've only used libraries that come internally with python.

There are many other libraries available that people have open-sourced and shared on PyPi

we will use **pip install** at the command line to install these packages

#### what is pip??

pip stands for "Pip Installs Packages"

It is the package manager for Python.

It helps you **download, install, upgrade, and remove** Python libraries from the Python Package Index (PyPI)
