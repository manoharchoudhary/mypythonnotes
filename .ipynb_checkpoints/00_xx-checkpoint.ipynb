```python
def prime(num):

    if num<2:
        return 0

    prime_ = [2]

    for i in range(3,num+1,2):
        for j in range(2,int(i**0.5) +1):
            if i%j==0:
                break
        else:
            prime_.append(i)

    return prime_, len(prime_)
                
```

```python
prime(100)
```
