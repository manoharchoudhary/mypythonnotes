```python
# to perform operation on file, first we need to open file
myfile = open('text.txt')
```

```python
# based on previous command we can calloff various method based on it:

#read file
myfile.read() 

# it return a giant string of everything in the text file
```

 \n (escape sequence): represent new line

```python
# if we try to read it again

myfile.read()
```

we get nothing in return, that is because: 
Imagine that there is cursor in the beginning of the file. when we read file for the first time, the cursor moves all the way to the end of the file. 
now if we try to read it again, because the cursor is at the end of the file. so there is nothing to read and we get an empty result.

###### we need to reset the cursor to the beginning of the file

```python
# reset the cursor
myfile.seek(0)

# .seek(0): reset the cursor at the beginning of the file
```

```python
#now we can read the file again
content = myfile.read()
```

```python
content
```

```python
myfile.seek(0)
```

### Reading, Writing, Appending modes of file

mode='r': is read only/n

mode='w: is write only(will overwrite files or creat new!)

mode='a': is append only(add more lines to file)

mode='r+': read and write

mode='w+': write and read (overwrite existing files)

```python
# read only
with open('text.txt',mode='r') as f:
    cont = f.readlines()
```

```python
cont
```

```python
# add more line
with open('text.txt',mode='a') as f:
    f.write('\nthis is fourth line')
```

```python
#read mode
with open('text.txt',mode='r') as f:
    cont = f.readlines()
    
cont
```

```python
#creating new file
# if file allready present then it will overwrite over the existing data
#if file do not exists then it will creat the new file

with open('abcd.txt',mode='w') as f:
    contents = f.write('i created this file')
```

```python
#read mode
with open('abcd.txt',mode='r') as f:
    print(f.readlines())
```
