### list:

```python
# return true if any num is even inside a list

def even_odd_list(num_list):
    for num in num_list:
        if num %2 ==0:
            return True
        else:
            pass
```

once a **return** is called the func is basically over. <br>
So **use "return" at the very end of the loop** or else you're going to break out of the loop without completing all the iterations.

```python
even_odd_list([1,2,3,4,5,6])
```

```python
even_odd_list([2,3])
```

```python
# our func do not return anything, if the number is false
even_odd_list([1,3,5,9])
```

```python
#return true if any number is even or return false

def even_odd_list(num_list):
    for num in num_list:
        if num %2 ==0:
            return True
        else:
            pass
    return False    
```

```python
even_odd_list([1,1,2])
```

```python
even_odd_list([1,1,3])
```

```python
#return all the even no. in a list

def even_odd_list1(num_list):
    even_num = []    #place holder
    
    for num in num_list:
        if num %2 ==0:
            even_num.append(num)
        else:
            pass
    return even_num
```

```python
even_odd_list1([1,2,3,4,5,6])
```

```python
even_odd_list1([1,3,5,9])
```

##### little but of tuple unpacking

```python


def odd_even(num_):
    odd_list = []
    even_list = []

    for num in num_:
        if num%2==0:
            even_list.append(num)
        else:
            odd_list.append(num)
    return (f"odd_list: {odd_list}"), (f"even_list: {even_list}")
   # return even_list
```

```python
odd_even([1,2,3,4,5,6,77])
```

```python
odd,even = odd_even([1,2,3,4,5,6,77])
```
