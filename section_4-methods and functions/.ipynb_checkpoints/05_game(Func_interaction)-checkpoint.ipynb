### game(3cups_guess the ball)

#### shuffling the cups

```python


def shuffled_func():
    cups = [0,'','']
    from random import shuffle
    shuffle(cups)
    return cups
```

```python
re = shuffled_func()
re
```

#### taking user input

```python
#taking user input

def user_input():
    result = int(input("Enter you're choice 1,2 or 3: "))

    if 1<= result<=3:
        return result-1
    else:
        print("Number is out of Bound.")
        user_input()
        
```

```python
user_input()
```

```python
def user_input_1():
    guess = ''
    while guess not in ['1','2','3']:
        guess = input("Enter you're choice 1,2 or 3: ")

    return int(guess)
```

```python
user_input_1()
```

#### comparing shuffling cups with the user input

```python
def checking_guess():

    cups = shuffled_func()
    guesses = user_input_1()

    if cups[guesses-1] == 0:
        print("Congrats you choosed wisely")
        print(cups)
    else:
        print("No good Guess, Try again")
        print(cups)
        checking_guess()
```
