```python
# creating function with def keyword

def say_hello():
    print('hello world')
```

snake caseing:-- using lower alphabets and underscore between words eg: your_name_is

most of the time we are going to use **return** keyword to return something out of a function

```python
#calling function

say_hello()
```

```python
# calling function without ()
say_hello
```

```python
# argument passing
def sayhello(name):
    print('hello ' + name)
```

```python
sayhello('marx')
```

```python
# setting up default argument
# if argument is not provided default value will get print

def sayhello(name = 'Nobuddy'):
    print ('hello ' + name)
```

```python
sayhello()
```

```python
sayhello('lily')
```

```python
result = sayhello('lily')
```

```python
result
```

```python
type(result)
```

#### using return func

```python
# return func

def add_num(num1,num2):
    return num1+num2
```

```python
add_num(3,10)
```

#### The advantage of using return instead of print is that :
#### It allows us to assign variable to the output of the function



```python
result = add_num(10,100)
```

```python
result
```

##### print VS result

```python
# can not assign func o/p to variable
def print_result(num1,num2):
    print(num1+num2)
```

```python
result5 = print_result(12,13)
```

```python
result5
```

```python
# can assign func o/p to variable
def print_return(num1,num2):
    return num1+num2
```
