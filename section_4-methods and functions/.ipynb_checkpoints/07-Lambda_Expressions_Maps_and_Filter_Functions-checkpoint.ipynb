## map

```python
# simple function that take a number and return a square
def square(num):
    return num**2
```

```python
list1 = [1,2,3,4,5,6]
```

Now, we want to apply square() function to each number in the list1,

```python
#one way to do this 
for num in list1:
    print(square(num))
```

Instead, we can take advantage of that **map function**

```python
map(square,list1)
```

```python
list(map(square,list1))  
# this applies the function to each element in the list
```

map applies function, to every element in the list

```python
for num in map(square,list1):
    print(num)
```

```python
# another example

def slicer(mystring):
    if len(mystring)%2==0:
        return "EVEN"
    else:
        return mystring[0]
    
```

```python
name = ["Andy","Eve","Sally"]
```

```python
list(map(slicer,name))
```

## filter

filter applies to a function that returns a boolean value (True or False)

```python
def check_even(num):
    return num%2==0   

#if num is even :---> True
#if num is odd :---> False
```

```python
check_even(10)
```

```python
check_even(9)
```

```python
list1 = [1,2,3,4,5,6]
```

What we could do is, use the **filter function**

```python
filter(check_even,list1)
```

```python
list(filter(check_even,list1))

#filter only return number which passes the function condition.
```

```python
#we can also iterate through the list

for x in filter(check_even,list1):
    print(x)
```

## Lambda Expression

Lambda expressions are a way to quickly create what are known as **anonymous functions**.

Basically, just one time use functions
that you don't even really name.

```python
# function that take in a num and return its square

#step1
def square(num):
    result = num**2
    return result
```

```python
#step2
def square(num):
    return num**2
```

```python
#step3
def square(num):return num**2
```

```python
# A lambda expression is also known as an anonymous function.
# because we only use one time 
# we don't actually give it a name,

# nor do we use the def keyword.
# So instead of a name or the def keyword,

# we replace that with the keyword, lambda.
```

```python
# and the return keyword is also removed
# because it's assumed that we're going to return whatever's on the other side of this colon(:)
```

```python
lambda num: num**2    #This is now a lambda expression.
```

```python
square1 = lambda num: num**2
```

```python
square1(8)
```

 lot of times, you're not really going to be doing

the "varibale name" is equal to some lamda expression.

Instead, you're going to be using it in **conjunction**

with other functions such as **map** and **filter**.

```python
list1 = [1,2,3,4,5,6]
```

#### map in lambda expression

# instead of defineing square function saperately we can use lambda expression

list(map(lambda num: num**2,list1))

```python
list(map(lambda num: num**2,list1))
```

#### filter in lambda expression

```python
# return only even numbers from the list1

list(filter(lambda num: num%2==0,list1))
```

##### keep in mind

not every single complex function

is going to be directly translated to a lambda expression.
