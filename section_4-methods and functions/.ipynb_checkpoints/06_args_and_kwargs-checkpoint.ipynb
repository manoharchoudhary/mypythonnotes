### *args: arguments 
### **kwargs: keyword arguments 

#### *args:
"*" is the actual keyword.  "args" is essentially a variable name it can be anything; eg: ***nums**, or ***jelly**.
but it is recommended that we only use "args" as variable name to go with the *

#### **kwargs
similarly:  " ** " is the actual keyword and "kwargs" is the varible name.It too can be anything  but it is recemmoneded to use "kwargs" to go with the "**"

#### *args: returns back a tuple
#### **kwargs: returns back a dictionary

### *args

```python
#retrun 5% of the sum a and b.

def myfunc(a,b):
    return sum((a,b))*0.05
```

```python
myfunc(50,60)
```

#####  a and b are example of positional argument 

now pass in more numbers or parameters 

```python
#one way to do so, is to increase the parameter in the function

def myfunc(a,b,c=0,d=0,e=0): #if you do not give value, default value will be taken
    return sum((a,b,c,d,e))*0.05
```

```python
myfunc(50,60,20)
```

```python
myfunc(50,60,20,30,50)
```

```python
myfunc(50,60,20,30,50,25)

#error: TypeError: myfunc() takes from 2 to 5 positional arguments but 6 were given
```

```python
# this is where we can use that *args 
# which will allows us to take an Arbitrary numbers of arguments

def myfunc(*args):
    print(args)
    return sum(args)*0.05
```

###### args actually look like a tuple

```python
myfunc(10,20,30)
```

```python
# we can give any number of arguments and it can take them with out any problem

## *args return backs a tuple
myfunc(10,20,30,40,50,60,1,56,40,15)
```

```python
# we can also use them for iteraction

def myfunc(*sam):
    for num in sam:
        print(num)
        
#args: is actually a variable so it can be anything
```

```python
myfunc(10,20,30,40,50,60,1,56,40,15)
```

### **kwargs : arbitrary number of key:value arguments

**kwargs: it builds a dictionary of key value pairs.

```python
def greet(**kwargs):     #kwargs can take any no. of key-value pairs
    for key,value in kwargs.items():
        print(f'{key} is {value}')
```

```python
greet(name = "marx", age =23, value = "none")
```

```python
#example

def myfunc(**kwargs):
    print(kwargs)    ## it is just a dictionary
```

```python
myfunc(fruit="mango",juice='coconut',vegies='potato')
```

```python
def myfunc(**kwargs):
    if "fruit" in kwargs.keys(): 
        #basically kwargs is a dictionary
        # fruit is a key value.
        print(f'my fruit is {kwargs["fruit"]}')
    else:
        print("no fruit")
    print("\n")
    print(kwargs) #it returns a dictionary
```

```python
myfunc(fruit = "mango",juice = 'coconut',vegies='potato')
```

```python
def myfunc(**kwargs):
    if "fruit" in kwargs: 
        print(f'my fruit is {kwargs["juice"]}')
    else:
        print("no fruit")
```

```python
myfunc(fruit = "mango",juice = 'coconut',vegies='potato')
```

```python
def myfunc(**kwargs):
    if "soup" in kwargs: 
        print(f'my fruit is {kwargs["juice"]}')
    else:
        print("no fruit")
```

```python
myfunc(fruit="mango",juice='coconut',vegies='potato')
```

##### Using both together

```python
#we can use both in combination 

def myfunc(*args,**kwargs):
    print(f'I would like {args[0]} {kwargs["food"]}')
    print("\n")
    print(args)
    print(kwargs)
```

```python
myfunc(20,30,50,food='orange',animal='dog',fruit='potato')
# the order should be the same 1st args than kwargs
```
