```python
def is_prime(n):
    if n <= 1:
        return False  # 0 and 1 are not prime
    if n <= 3:
        return True   # 2 and 3 are prime
    
    # Eliminate multiples of 2 and 3 quickly
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Check divisibility from 5 up to âˆšn
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage

```

```python
print(is_prime(29))  # True
print(is_prime(30))  # False
```
