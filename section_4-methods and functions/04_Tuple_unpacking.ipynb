### Tuple unpacking with python

```python
stock_prices = [('apple' , 200),('google' ,400),('msf',100)]
```

```python
for item in stock_prices:
    print(item)
```

```python
# tuple unpacking 

for comp,prices in stock_prices:
    print(comp)
```

```python
for comp,prices in stock_prices:
    print(prices)
```

```python
#list of workers in shop one

shop1 = [('abby',100),('billy',400),('cassie',800)]
```

```python
# find emp who worked most hours

def emp_check(work_hours):
    current_max = 0
    emp_of_month = ''
    
    for emp,hour in work_hours:
        if current_max < hour:
            current_max = hour
            emp_of_month = emp
        else:
            pass
            
    return (emp_of_month,current_max)
```

```python
emp_check(shop1)
```

```python
# as our func is returning a tuple, Hence
result = name,hour = emp_check(shop1)
```
