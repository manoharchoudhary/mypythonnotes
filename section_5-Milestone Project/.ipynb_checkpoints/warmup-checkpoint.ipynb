```python
# Validating User Input

def user_choice():
    
    while True:
        choice = input("Enter a number between(1-10): ")

        if not choice.isdigit():
            print("Please enter a number")
            continue 

        if int(choice) not in range(1,11):
            print("Number not in range")
            continue
            
        return int(choice)
```

```python
user_choice()
```
